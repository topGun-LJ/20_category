```python
# 定义父文件夹的路径
folder_path = "20_newsgroups"
# 获取该文件夹下的所有子文件夹名称，并存储在 subfolders 列表中
subfolders = [f.name for f in os.scandir(folder_path) if f.is_dir()]
# 对子文件夹名称列表进行排序
subfolders.sort()
# 创建一个空字典，用于存储映射关系
mapping_dict = {}
# 使用 enumerate 函数遍历子文件夹列表，将子文件夹名称映射到数字 0-19
for i, subfolder in enumerate(subfolders,start = 1):
    mapping_dict[subfolder] = i
print(mapping_dict)
# 将映射字典存储为 pickle 文件
with open("folder_mapping.pickle", "wb") as f:
    pickle.dump(mapping_dict, f)

mmap = {}
with open("folder_mapping.pickle", "rb") as f:
    mmap = pickle.load(f)
    
print(mmap)
```

```python
# 读取存储映射关系的 pickle 文件
with open("folder_mapping.pickle", "rb") as f:
    mapping_dict = pickle.load(f)


# 遍历映射关系中的键值对
for old_folder_name, new_folder_name in mapping_dict.items():
    old_folder_path = os.path.join("20_newsgroups", old_folder_name)
    new_folder_path = os.path.join("20_newsgroups", str(new_folder_name))  # 修正：加入父文件夹路径
    # 检查旧文件夹是否存在
    if os.path.exists(old_folder_path):  # 修正：使用完整的旧文件夹路径
        try:
            # 重命名文件夹
            os.rename(old_folder_path, new_folder_path)
            print(f"Renamed {old_folder_path} to {new_folder_path}")
        except Exception as e:
            print(f"Failed to rename {old_folder_path} to {new_folder_path}: {e}")
```
